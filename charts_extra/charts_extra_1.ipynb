```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


#import data and create dataframe
data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data' 
data_column_headers = ['sepal_length','sepal_width','petal_length','petal_width', 'class']
df = pd.read_csv(data_url, names = data_column_headers)

# if we want to categorize into numeric terms our classes
df['class'] = pd.Categorical(df['class'])
df['class_code'] = df['class'].cat.codes

# let's take a look at our data
df.head()

#figure
fig, ax1 = plt.subplots()
fig.set_size_inches(16, 10)

#labels
ax1.set_xlabel('Sepal Length')
ax1.set_ylabel('Sepal Width')
ax1.set_title('Relationship Between Sepal Length and Sepal Width in Irises')

#setting our color map setting our dot size
plt.scatter( df['sepal_length'], df['sepal_width'] ,\
             c=df['sepal_width'], cmap = 'viridis',\
             s = pow(df['sepal_width'],4),\
             alpha =.5, edgecolor="blue") 

cbar = plt.colorbar()
cbar.set_label('Sepal Width')
```
