```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import ListedColormap

"""
Creates a heatmap that highlights outliers
"""

color_range = cm.get_cmap('viridis', 256) #cmap is a colormap and it comes in different styles (viridis,spring,winter,inferno,etc.)
custom_color_range = color_range(np.linspace(0, 1, 256)) # create a color_range
outlier_color = np.array([256/256, 0/256, 0/256, 1]) # outlier color defined by RGB (Red/256,Green/256,Blue/256,[0-1] for brightness)
outlier_percentage = .25 # range of [0-1] 0=0% 1=100%
custom_color_range[:int(256*outlier_percentage), :] = outlier_color
newcmp = ListedColormap(custom_color_range)

data=np.random.randn(90,90)

def plot_heatmap(colormaps):   
    n = len(colormaps)
    fig, axs = plt.subplots(1, n, figsize=(n * 2 + 2, 3),
                            constrained_layout=True, squeeze=False)
    for [ax, cmap] in zip(axs.flat, colormaps):
        psm = ax.pcolormesh(data, cmap=cmap, rasterized=True, vmin=-4, vmax=4)
        fig.colorbar(psm, ax=ax)
    plt.title(f"Heatmap with a constraint of {outlier_percentage*100} %") 
    plt.show()

plot_heatmap([newcmp])

#plt.imshow(np.random.random((30,30)), cmap = 'viridis') #inferno, winter, spring 
```
