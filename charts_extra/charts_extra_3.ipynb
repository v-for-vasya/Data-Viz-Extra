```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

n=100 #create n paths

df = pd.DataFrame(np.random.randn(365, n), index=pd.date_range('1/1/2000', periods=365), columns=np.arange(n))
df = df.cumsum() # sums up our random numbers

colors_beautiful=['#fe1895','#fc0000','#fc9e0c','#fdf000','#79d000','#01d2d4']
colors_range=['0.1', '0.12', '0.14','0.16','0.18','0.2', '0.22', '0.25', '0.3', '0.35', '0.4', '0.45', '0.5']

plt.style.use('dark_background') #ggplot, seaborn, seaborn-whitegrid
df.plot(legend=False, color=colors_beautiful, linewidth=.4)
plt.title(f"Our Monte Carlo Simulation with {n} paths")
plt.show()
```
